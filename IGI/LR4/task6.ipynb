```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
titanic = pd.read_csv("https://bit.ly/kaggletrain")
titanic.shape
```

```python
titanic.set_index("PassengerId", inplace=True)
titanic.head(5)
```

```python
plt.figure(figsize=(8, 6))
sizes = titanic["Sex"].value_counts()
plt.pie(sizes, labels=sizes.index, autopct="%1.1f%%")
plt.title("All passengers")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
survived = titanic[titanic["Survived"] == 1]
sizes = survived["Sex"].value_counts()
plt.pie(sizes, labels=sizes.index, autopct="%1.1f%%")
plt.title("Survived passengers")
plt.show()
```

```python
sns.scatterplot(data=titanic, x="Fare", y="Age",hue="Sex", style="Sex")
```

```python
sns.kdeplot(data=titanic, x="Fare", y="Age",hue="Sex") # I suppose so
```

```python
men_survived = titanic.loc[(titanic["Sex"] == "male") & (titanic["Survived"] == 1) & (titanic["Age"] > 45)]
# 12
men_survived.plot(kind="bar")
plt.title("Survival of Men Over 45")
plt.xlabel("Categories")
plt.ylabel("Count")
plt.show()
men_survived
```

```python
sns.histplot(data=titanic, x="Fare")
```

```python
corr = titanic.corr(numeric_only=True)
sns.heatmap(corr, annot=True,fmt=".3f",linewidths=3, linecolor="red", cmap="crest")
plt.title("Matrix correlation")
plt.show()
```
